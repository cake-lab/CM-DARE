```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
plt.style.use('grayscale')
from matplotlib.font_manager import FontProperties
font0 = FontProperties()
def gen_font(weight, size):
    font1=font0.copy()
    font1.set_family('sans-serif')
    font1.set_weight(weight)
    font1.set_size(size)
    return font1
```

```python
res15k80 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/res15k80/data.csv')
res32k80 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/res32k80/data.csv')
shakesmallk80 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/shakesmallk80/data.csv')
shakebigk80 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/shakebigk80/data.csv')
res15p100 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/res15p100/data.csv')
res32p100 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/res32p100/data.csv')
shakesmallp100 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/shakesmallp100/data.csv')
shakebigp100 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/shakebigp100/data.csv')
res15v100 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/res15v100/data.csv')
res32v100 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/res32v100/data.csv')
shakesmallv100 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/shakesmallv100/data.csv')
shakebigv100 = pd.read_csv('/Users/ozymandias/desktop/experiment_results/speed_data_base/shakebigv100/data.csv')
# print res15k80
res15k80 = res15k80.drop(res15k80.index[[40,41]])
res32k80 = res32k80.drop(res32k80.index[[40,41]])
shakesmallk80 = shakesmallk80.drop(shakesmallk80.index[[40,41]])
shakebigk80 = shakebigk80.drop(shakebigk80.index[[40,41]])
res15p100 = res15p100.drop(res15p100.index[[0,39,40]])
res32p100 = res32p100.drop(res32p100.index[[0,39,40]])
shakesmallp100 = shakesmallp100.drop(shakesmallp100.index[[0,39,40]])
shakebigp100 = shakebigp100.drop(shakebigp100.index[[0,39,40]])
res15v100 = res15v100.drop(res15v100.index[[0,39,40]])
res32v100 = res32v100.drop(res32v100.index[[0,39,40]])
shakesmallv100 = shakesmallv100.drop(shakesmallv100.index[[0,39,40]])
shakebigv100 = shakebigv100.drop(shakebigv100.index[[0,39,40]])
```

```python
fig, ax = plt.subplots(figsize=(3.3,1.5))
font = gen_font('light',6)
font2 = gen_font('light',5)
# fig.tight_layout()
ax = res15k80.plot(ax=ax, y="Value",x="Step", label='ResNet-15',linestyle=':',lw=0.8)
ax = res32k80.plot(ax=ax, y="Value",x="Step", label='ResNet-32',linestyle='--',lw=0.8)
ax = shakesmallk80.plot(ax=ax, y="Value",x="Step", label='Shake Shake Small',linestyle='-.',lw=0.8)
ax = shakebigk80.plot(ax=ax, y="Value",x="Step", label='Shake Shake Big',linestyle='-',lw=0.8)
ax.set_ylim(0, 11)
plt.yticks(np.arange(0, 12, 2),fontproperties=font)
ax.set_xlim(0, 4000)
plt.xticks(fontproperties=font)
plt.setp(ax.spines.values(), color='lightgray')
plt.ylabel("Speed (steps/second)", fontproperties=font)
plt.xlabel("Number of Steps", fontproperties=font)
plt.legend(loc='upper center',ncol=2,bbox_to_anchor=(0.5, 1.25),prop=font2, framealpha=0)
# plt.show()
fig.savefig("/Users/ozymandias/Desktop/one_k80_cluster_training_speed_time_series.pdf", bbox_inches='tight')
```
